```python
df['Average'] = df[['Maths','Physics', 'Computer']].mean(axis=1).round(2)
df
```

```python
def grades(marks):
    if marks>90:
        return 'A'
    elif marks>80:
        return 'B'
    elif marks>70:
        return 'C'
    elif marks>70:
        return 'D'
    else: 
        return "Fail"
```

```python
df['Grade'] = df['Percentage'].apply(grades)
df
```
